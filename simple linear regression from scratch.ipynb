```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating dataset
x= np.arange(1,100,1)
y=[]
for i in range (len(x)):
    y.append(x[i]+float(np.random.uniform(-10,10)))
```

```python
plt.scatter(x,y)
plt.show()
```

```python
# Gradient Descent
def gradient_descent(x,y,m,c,lr=0.0001):
    m_grad=0
    c_grad=0
    n=len(x)
    for i in range(len(x)):
        m_grad+=-(2/n)*(y[i]-(m*x[i]+c))*x[i]
        c_grad=-(2/n)*(y[i]-(m*x[i]+c))
    
    m=m-lr*m_grad
    c=m-lr*c_grad
    return m,c
    
```

```python
# Creating plots
def plot_(epoch):
        plt.scatter(x,y)
        plt.plot(range(100),[m*x+c for x in range(100)],color='g')
        plt.text(x=1,y=90,s='epoch={}'.format(epoch))
        plt.show()
```

```python
# Visulaizing Best fit line
epoch=10
m=0
c=0
plot_(0)
print('m={} c={}'.format(m,c))
for i in  range (1,epoch+1):
    m,c=gradient_descent(x,y,m,c)
    if i%2==0:
        plot_(i)
        print('m={} c={}'.format(m,c))
        
```

```python
error=0
for i in range (len(x)):
    error+=(y[i]-(m*x[i]+c))**2
rmse= np.sqrt(error/len(x))
print('rmse={}'.format(rmse))
```
